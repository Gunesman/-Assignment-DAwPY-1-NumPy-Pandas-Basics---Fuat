```python
df['mass'] = df['mass'].str.replace(',', '', regex=True).astype('float')
df['mass']
```

```python
def categorize_mass(mass):
    if pd.isna(mass):
        return 'NaN'  # Eksik değerler
    elif mass >= 100:
        return 'High_mass'
    elif 50 < mass < 100:
        return 'Normal_mass'
    else:
        return 'Low_mass'

df['mass_cat'] = df['mass'].apply(pd.to_numeric, errors='coerce').apply(categorize_mass)
df['mass_cat'].value_counts()
```

```python
# with seaborn barplot

```

```python
# Mass categories with seaborn barplot 

def categorize_mass(mass):
    if pd.isna(mass):
        return 'NaN'  
    elif mass >= 100:
        return 'High_mass'
    elif 50 < mass < 100:  
        return 'Normal_mass'
    else:
        return 'Low_mass'

# Apply categorization
df['mass_cat'] = df['mass'].apply(pd.to_numeric, errors='coerce').apply(categorize_mass)

# Filter out NaN values
filtered_mass = df[df['mass_cat'] != 'NaN']

# Get category order
mass_order = filtered_mass['mass_cat'].value_counts().index

# Plot barplot
sns.barplot(
    x=filtered_mass['mass_cat'].value_counts().index,
    y=filtered_mass['mass_cat'].value_counts().values,
    order=mass_order,
    palette="deep"
)

# Title and labels
plt.title('Mass Groups', fontsize=20, color='blue')
plt.xlabel('mass_cat', fontsize=12)

# Add values on bars
mass_counts = filtered_mass['mass_cat'].value_counts()
for i, value in enumerate(mass_counts.values):
    plt.text(i, value + 0.25, str(value), ha='center', fontsize=12)

# Adjust axis labels
plt.xticks(fontsize=12, weight='normal')
plt.yticks(fontsize=12);
```

```python
# Mass groups with a light blue background
plt.figure(figsize=(10, 6))  # Set figure size

# Categorize mass
def categorize_mass(mass):
    if pd.isna(mass):
        return 'NaN'  
    elif mass >= 100:
        return 'High_mass'
    elif 50 < mass < 100:
        return 'Normal_mass'
    else:
        return 'Low_mass'

df['mass_cat'] = df['mass'].apply(pd.to_numeric, errors='coerce').apply(categorize_mass)

# Filter NaN values
filtered_mass = df[df['mass_cat'] != 'NaN']

# Sort categories
mass_order = filtered_mass['mass_cat'].value_counts().index

# Set light blue background
plt.gca().set_facecolor("#ADD8E6")

# Create barplot
sns.barplot(
    x=filtered_mass['mass_cat'].value_counts().index,
    y=filtered_mass['mass_cat'].value_counts().values,
    order=mass_order,
    palette=["#F4A582", "#E08214", "#B2182B"]  # Pastel colors
)

# Title and labels
plt.title('Mass Groups', fontsize=18, color='blue')
plt.xlabel('mass_cat', fontsize=12)
plt.ylabel('count', fontsize=12)

# Add values on bars
mass_counts = filtered_mass['mass_cat'].value_counts()
for i, value in enumerate(mass_counts.values):
    plt.text(i, value + 0.25, str(value), ha='center', fontsize=12, color='red')

# Adjust axis labels
plt.xticks(fontsize=12, weight='normal', rotation=45)
plt.yticks(fontsize=12);
```
